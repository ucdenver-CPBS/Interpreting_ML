### Load Dataset

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
df_iris = pd.DataFrame(iris.data,columns=iris.feature_names)
df_iris['target'] = pd.Series(iris.target)

X = df_iris.drop(['target'], axis=1)
y = df_iris['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=500)
model.fit(X_train, y_train)
```

### Permutation Importances (EIL5)

```python
import eli5
from eli5.sklearn import PermutationImportance
perm = PermutationImportance(model, random_state=1).fit(X_train, y_train)
eli5.show_weights(perm, feature_names = X_train.columns.tolist(), top=5)
```

```python
from eli5 import show_prediction

## Show prediction of features in individual 
show_prediction(model, X_test.iloc[7], feature_names = X_train.columns.tolist(), 
                show_feature_values=True)
```

### PDP

```python
from matplotlib import pyplot as plt
from pdpbox import pdp, get_dataset, info_plots

pdp_width = pdp.pdp_isolate(model=model, dataset=X_train, model_features=X_train.columns.tolist(), feature='petal width (cm)')

pdp.pdp_plot(pdp_width, 'petal width (cm)')
plt.show()
```

```python
pdp_interaction = pdp.pdp_interact(model=model, dataset=X_train, model_features=X_train.columns.tolist(), features=['petal width (cm)', 'petal length (cm)'])
pdp.pdp_interact_plot(pdp_interact_out=pdp_interaction, feature_names=['Petal Width', 'Petal Length'], plot_type='contour')
plt.show()
```

### LIME

```python
import lime.lime_tabular
explainer = lime.lime_tabular.LimeTabularExplainer(X_train.values,
                                                   feature_names=X_train.columns.values.tolist(),
                                                   class_names=y_train.unique())

predict_fn = lambda x: model.predict_proba(x).astype(float)
```

```python
exp = explainer.explain_instance(X_test.values[7], predict_fn, num_features=3)
exp.show_in_notebook(show_all=False)
```

### SHAP

```python
## Switch to Wisconson Breast Cancer Dataset (Binary Classification)
bc = datasets.load_breast_cancer()
df_bc = pd.DataFrame(bc.data,columns=bc.feature_names)
df_bc['target'] = pd.Series(bc.target)

X = df_bc.drop(['target'], axis=1)
y = df_bc['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=500)
model.fit(X_train, y_train)
```

```python
import shap
example_1 = X_test.iloc[7]
```

```python
explainer = shap.TreeExplainer(model, X_train)
shap_values = explainer.shap_values(X_train)
shap.summary_plot(shap_values[1], X_train)
```

```python
shap.dependence_plot('worst perimeter', shap_values[1], X_train, interaction_index='worst concave points')
```

```python
shap_values = explainer.shap_values(example_1)
shap.initjs()
shap.force_plot(explainer.expected_value[1], shap_values[1], example_1)
```

### mlxtend

```python
from mlxtend.plotting import plot_decision_regions
from mlxtend.classifier import EnsembleVoteClassifier

import itertools

import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt


from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier

## Reduce to top 2 features
X_train_ml = X_train[['worst perimeter', 'worst concave points']].values
y_train_ml = y_train.values
```

```python
clf1 = LogisticRegression(random_state=1)
clf2 = RandomForestClassifier(random_state=1)
clf3 = GaussianNB()
eclf = EnsembleVoteClassifier(clfs=[clf1, clf2, clf3], weights=[1,1,1])
value=1.5
width=0.75
gs = gridspec.GridSpec(2,2)
fig = plt.figure(figsize=(10,8))
labels = ['Logistic Regression', 'Random Forest', 'Naive Bayes', 'Ensemble']
for clf, lab, grd in zip([clf1, clf2, clf3, eclf],
                         labels,
                         itertools.product([0, 1], repeat=2)):
                         
    clf.fit(X_train_ml, y_train_ml)
    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=X_train_ml, y=y_train_ml, clf=clf)
    plt.title(lab)
```
